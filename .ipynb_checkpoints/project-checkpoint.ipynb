# First try: I use the method we learn in class to regress the SalePrice and use gradient descent to decrease the least square.

Because the four factors I choose are very different, that is to say the difference of their magnitudes is really big, I need to normalize the input data first.

```python
# a normalization function 
def normalize(data,data_mean,data_std):
    normalized_data = (data - data_mean)/data_std
    return normalized_data
```

```python
x_means = np.mean(trainx,axis = 0)
x_stds = np.std(trainx,axis = 0)
x_normed = normalize(trainx,x_means,x_stds)
```

```python
def predict(x,w):
    return w[0]+w[1]*x[0]+w[2]*x[1]+w[3]*x[2]+w[4]*x[3]
def least_square(w):
    cost=0
    for i in range(len(trainx)):
        ls=(predict(x_normed[i],w)-trainy[i])**2
        cost=cost+ls
    return cost    
def gradient_descent(g,w,alpha,max_its,beta,version):    
    # flatten the input function, create gradient based on flat function
    g_flat, unflatten, w = flatten_func(g, w)
    grad = compute_grad (g_flat)

    # record history
    w_hist = []
    w_hist.append(unflatten(w))

    # start gradient descent loop
    z = np.zeros((np.shape(w)))      # momentum term
    
    # over the line
    for k in range(max_its):   
        # plug in value into func and derivative
        grad_eval = grad(w)
        grad_eval.shape = np.shape(w)

        ### normalized or unnormalized descent step? ###
        if version == 'normalized':
            grad_norm = np.linalg.norm(grad_eval)
            if grad_norm == 0:
                grad_norm += 10**-6*np.sign(2*np.random.rand(1) - 1)
            grad_eval /= grad_norm
            
        # take descent step with momentum
        z = beta*z + grad_eval
        w = w - alpha*z

        # record weight update
        w_hist.append(unflatten(w))

    return w_hist
```

```python
w_init= 100*np.random.randn(5,1)
# parameters of gradient descent
alpha =10**(4) ;  max_its = 75; beta = 0; 
w_history = gradient_descent(least_square,w_init,alpha,max_its,beta,version = 'normalized')
```

```python
# plot cost function history
demo = util.Visualizer()
cost_history = [least_square(v) for v in w_history]
histories = [cost_history]
demo.compare_regression_histories(histories)
```

```python
np.argmin(cost_history)
best_w=w_history[-1]
```

```python
def price_predict(x):
    w=best_w
    return w[0]+w[1]*x[0]+w[2]*x[1]+w[3]*x[2]+w[4]*x[3]
```

```python
finalprice=[]
for i in range(len(trainx)):
    price=price_predict(trainx[i])
    finalprice.append(price)
```

# Second try: Gradient Boosting Regression which is an useful method to solve the data prediction problem

```python
gbr =GBR(n_estimators=3000, learning_rate=0.5,max_depth=4, max_features='sqrt',min_samples_leaf=15, min_samples_split=10, loss='huber', random_state =5)
gbr.fit(trainx, trainy)
gbr_preds = gbr.predict(trainx)
```

```python
trainy=trainy.reshape(1453,1)
gbr_preds=gbr_preds.reshape(1453,1)
gbr_residuals = abs(trainy - gbr_preds)
ch_residuals = abs(trainy - finalprice)
```

# Compare the result of two method, it is easy to find that the second way is much better.

```python
x=range(len(ch_residuals ))
plt.figure(figsize=(15,10))
plt.plot(x,ch_residuals,'b',label='first method')
plt.plot(x,gbr_residuals,'r',label='second method')
plt.legend(loc='upper right')
```

Compared with the first method, the second method's error can be ignored. So I use the gradient boosting regressor to predict the test data.

```python
test_preds = gbr.predict(testx)
```
